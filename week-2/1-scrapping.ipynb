# **Scrapping Google Play Review Jenius App**

Pada tahap pengumpulan data, digunakan teknik web scrapping pada platform Google Play Store dengan target aplikasi perbankan digital Jenius. Ekstraksi ulasan pengguna dilakukan menggunakan library **`google-play-scraper`** dengan memanfaatkan fungsi **`reviews_all`** yang memungkinkan pengambilan data secara menyeluruh tanpa batasan jumlah. Konfigurasi scraping diarahkan pada identifikasi paket aplikasi **`com.spotify.music`** dengan parameter bahasa ditetapkan pada Bahasa Indonesia (**`'id'`**) guna memfokuskan perolehan data pada ulasan berbahasa Indonesia. Keseluruhan ulasan yang tersedia diambil menggunakan metode pengurutan berdasarkan ulasan terbaru **`(Sort.NEWEST)`** untuk menjamin relevansi dataset dengan kondisi aktual aplikasi pada saat penelitian dilakukan. Tahapan ini berfungsi sebagai fondasi pembangunan dataset primer yang akan menjadi bahan analisis pada proses selanjutnya.

## Instalasi Library

```python
from google_play_scraper import app, reviews_all, Sort
from datetime import datetime
import pandas as pd
import numpy as np
```

## Detail Information Jenius App

```python
# 1. Mengambil detail aplikasi menggunakan App ID Spotify
result = app(
    'com.spotify.music',
    lang='id',
    country='id'
)

# 2. Konversi Timestamp ke Format Tanggal Manusia
dt_object = datetime.fromtimestamp(result['updated'])
tanggal = dt_object.strftime('%d %B %Y, %H:%M WIB')

# 3. Menampilkan informasi spesifik terkait update
print(f"Nama Aplikasi     : {result['title']}")
print(f"Versi Terbaru     : {result['version']}")
print(f"Terakhir Diupdate  : {tanggal}")

# 4. Menampilkan dalam bentuk DataFrame
df_detail = pd.DataFrame([result])
df_detail['updated_formatted'] = tanggal
display(df_detail[['title', 'version', 'updated_formatted', 'installs']])
```

##Scrapping Jenius App Review

```python
# 1. Proses scrap
spotify_reviews_id = reviews_all(
    'com.spotify.music',
    sleep_milliseconds=0,
    lang='id',
    country='id',
    sort=Sort.NEWEST,
)

print(f'✅ Scraping selesai! Total ulasan terkumpul: {len(spotify_reviews_id):,}')

# 2. Menyimpan ke DataFrame
df_spotify_id = pd.DataFrame(np.array(spotify_reviews_id), columns=['content'])
df_spotify_id = df_spotify_id.join(pd.DataFrame(df_spotify_id.pop('content').tolist()))

print(f'\nJumlah baris  : {df_spotify_id.shape[0]:,}')
print(f'Jumlah kolom  : {df_spotify_id.shape[1]}')

df_spotify_id.head()
```

```python
# 3. Simpan ke file csv
output_path = "../out/rawdata_spotify_id.csv"
df_spotify_id.to_csv(output_path, index=False, encoding="utf-8")
print(f'\n✅ Data berhasil disimpan ke: {output_path}')
```
